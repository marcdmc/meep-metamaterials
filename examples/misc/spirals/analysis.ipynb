```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt
import sys, os
sys.path.append('../../..')
from meep_metamaterials import metamaterials as mm
from meep_metamaterials.retrieval import retrieval
```

```python
freqs = np.linspace(0.2, 1, 50)
wvl = 1/freqs
s11 = np.loadtxt('s11_test3.txt', dtype=np.complex128)
s21 = np.loadtxt('s21_test3.txt', dtype=np.complex128)
params = retrieval.eff_parameters(freqs, 0.18, s11, s21, plot_branches=True, continuity=True, branch=0, th=10)
```

```python
plt.plot(wvl, np.abs(s11)**2, label='|S11|^2')
plt.plot(wvl, np.abs(s21)**2, label='|S21|^2')
plt.xlabel('Wavelength (um)')
plt.legend()
```

```python
plt.plot(freqs*3e8*1e6/1e12, np.real(params['eps']), 'b', label='Re(eps)')
plt.plot(freqs*3e8*1e6/1e12, np.real(params['mu']), 'r', label='Re(mu)')
plt.xlabel('Frequency (THz)')
plt.grid()
plt.legend()
```
