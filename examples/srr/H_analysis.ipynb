```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt
import sys, os
sys.path.append('../..')
from meep_metamaterials import metamaterials as mm
from meep_metamaterials.retrieval import retrieval
```

```python
freqs = np.linspace(0.3, 0.7, 50)
wvl = 1/freqs
a = 0.45
l = 0.32 # Now the thickness is the side of the resonator

s11 = np.loadtxt('tmp/H/s11_H.txt', dtype=np.complex128)
s21 = np.loadtxt('tmp/H/s21_H.txt', dtype=np.complex128)
params = retrieval.eff_parameters(freqs, l, s11, s21, plot_branches=True, branch=0)

# With two layers
s11_2 = np.loadtxt('tmp/H/s11_2layers_H.txt', dtype=np.complex128)
s21_2 = np.loadtxt('tmp/H/s21_2layers_H.txt', dtype=np.complex128)
params_2 = retrieval.eff_parameters(freqs, a+l, s11_2, s21_2, plot_branches=True, branch=0, continuity=True)

# With three layers
s11_3 = np.loadtxt('tmp/H/s11_3layers_H.txt', dtype=np.complex128)
s21_3 = np.loadtxt('tmp/H/s21_3layers_H.txt', dtype=np.complex128)
params_3 = retrieval.eff_parameters(freqs, 2*a+l, s11_3, s21_3, plot_branches=True, branch=0, continuity=False)
```

```python
plt.figure()
plt.plot(freqs*3e8*1e6/1e12, np.abs(s11)**2, 'b', label='S11')
plt.plot(freqs*3e8*1e6/1e12, np.abs(s21)**2, 'r', label='S21')
plt.grid()
plt.legend()
plt.xlabel('Frequency (THz)')

plt.figure()
plt.plot(freqs*3e8*1e6/1e12, np.real(params['eps']), 'b', label='Re($\epsilon$)')
plt.plot(freqs*3e8*1e6/1e12, np.real(params['mu']), 'r', label='Re($\mu$)')
plt.grid()
plt.legend()
plt.xlabel('Frequency (THz)')
```

```python
plt.plot(freqs, np.real(params['eps']), 'b', alpha=.3)
plt.plot(freqs, np.real(params['mu']), 'r', alpha=.3)

plt.plot(freqs, np.real(params_2['eps']), 'b', label='Re($\epsilon$)')
plt.plot(freqs, np.real(params_2['mu']), 'r', label='Re($\mu$)')

plt.grid()
plt.legend()
plt.xlabel('Frequency (THz)')
plt.title('Two layers')
```

```python
plt.plot(freqs, np.real(params['eps']), 'b', alpha=.2)
plt.plot(freqs, np.real(params['mu']), 'r', alpha=.2)

plt.plot(freqs, np.real(params_2['eps']), 'b', alpha=.3)
plt.plot(freqs, np.real(params_2['mu']), 'r', alpha=.3)

plt.plot(freqs, np.real(params_3['eps']), 'b')
plt.plot(freqs, np.real(params_3['mu']), 'r')

plt.grid()
plt.legend()
plt.xlabel('Frequency (THz)')
plt.title('Three layers')
```
